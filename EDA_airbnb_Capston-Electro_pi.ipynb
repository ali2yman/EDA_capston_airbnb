<B>Find the most highly rated neighborhoods<B>

```python
top_rated = df.groupby('neighbourhood')['number_of_reviews'].sum().sort_values(ascending=False)
top_rated = top_rated.head(10)
# Plot the results using a bar plot
plt.figure(figsize=(12, 7))
top_rated.plot(kind='bar', color = "r")
plt.xlabel('Neighborhood')
plt.ylabel('Total Number of Reviews')
plt.title('Top Rated Neighborhoods')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

<b>Explore the neighbourhood distribution using `count plots` to see which neighbourhoods have the `highest number of listings`<B>
